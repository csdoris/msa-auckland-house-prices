# Auckland House Prices Analysis

## Read data collected from part-1-data-collection

```r
df<-read.csv('convert-land-area-to-float.csv')
df
```

## Descriptive statistics

```r
summary(df)
```

## One-way ANOVA - CV ~ Suburbs

```r
fit1<-aov(CV~as.factor(Suburbs),data=df)
summary(fit1)
```

```r
drop1(fit1,~.,test="F")
```

## One-way ANOVA - CV ~ NZDep2018

```r
fit2<-aov(CV~as.factor(NZDep2018),data=df)
summary(fit2)
```

```r
drop1(fit2,~.,test="F")
```

### Pairwise comparisons using Tamhane's T2-test for unequal variances

```r
library("PMCMRplus")
summary(T2 <- tamhaneT2Test(df$CV, as.factor(df$NZDep2018)))
```

## Two-way ANOVA

```r
fit3 <- aov(CV ~ as.factor(Suburbs)+as.factor(NZDep2018)+ as.factor(Suburbs):as.factor(NZDep2018), data=df)
summary(fit3) 
```
